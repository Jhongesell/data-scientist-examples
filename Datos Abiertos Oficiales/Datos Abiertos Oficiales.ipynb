# Obtener los datos

```python
import pandas as pd
```

```python
covid_df = pd.read_csv("https://docs.google.com/spreadsheets/d/16-bnsDdmmgtSxdWbVMboIHo5FRuz76DBxsz_BbsEVWA/export?format=csv&id=16-bnsDdmmgtSxdWbVMboIHo5FRuz76DBxsz_BbsEVWA&gid=0")
```

```python
covid_df
```

## Reemplazando los "NaN"

```python
# Obtengo las columnas
columnas = covid_df.columns.values.tolist()
```

```python
# Reemplazo según el tipo de dato, si es un "object" le modifico el NaN por "Sin información"
# Si es un float64 o int64, lo modifico por un 0 (de forma automática se coloca el 0.0 o 0 correspondiente a cada columna)
for i in columnas:
    if str(covid_df[i].dtypes) == "object":
        covid_df[i] = covid_df[i].replace(float("nan"), "Sin información")
    elif str(covid_df[i].dtypes) == "float64" or "int64":
        covid_df[i] = covid_df[i].replace(float("nan"), 0)
```

```python
covid_df
```

# Recopilando las provincias en un diccionario

```python
lista_provincias = ["Buenos Aires", "CABA", "Catamarca", "Chaco",
                    "Chubut", "Córdoba", "Corrientes",
                    "Entre Ríos", "Formosa", "Jujuy",
                    "La Pampa", "La Rioja", "Mendoza",
                    "Misiones", "Neuquén", "Río Negro",
                    "Salta", "San Juan", "San Luis",
                    "Santa Cruz", "Santa Fe", "Santiago del Estero",
                    "Tierra del Fuego", "Tucumán"]

# Para mostrar el dataframe solo es necesario llamar al key
# df_provincias["Nombre de la provincia"]

# Si queremos saber por ejemplo las filas y columnas
# df_provincias["Nombre de la provincia"].shape
```

```python
df_provincias = {provincia:covid_df[covid_df["osm_admin_level_4"] == provincia] for provincia in lista_provincias}
```

```python
# Ejemplo con una provincia

df_provincias["Salta"]
```

# Obteniendo datos más importantes

### Si bien tenemos en el diccionario filtrado con todos los datos de cada provincia, si queremos tener datos más útiles a disposición podemos crearlos en una variable aparte con el nombre de cada provincia

| Nombre de columna         | Descripción                                                        |
|----------------------|-------------------------------------------------------------------|
| fecha                | fecha a la que corresponde los datos                              |
| dia_inicio           | cant dias desde el inicio del caso 1                              |

| nue_casosconf_diff           | nuevos casos infectados del dia del lugar especificado en las columnas osm admin level                                          |
| tot_fallecidos             | total de fallecidos. Columna que sumariza fila a fila el total de fallecidos                                             |
| nue_fallecidos_diff                | nuevos casos fallecidos del dia                                                      |
| tot_recuperados                  | total acumulado de casos recuperados.                                                       |

| tot_test             | total acumulado de tests |

| informe_link | URL de acceso al informe de donde sale el dato                                            |
| transmision_tipo | Region shapefile in WKT                                            |

| covid19argentina_admin_level_4 | formato provincia requerido por  necesario covid19argentina.com                                            |


### Los datos que más nos interesarían por provincia son fecha, dia_inicio, dia_cuarentena_dnu260, nue_casosconf_diff, nue_fallecidos_diff, informe_link, observacion


```python
# Por lo tanto, en una variable aparte si nos interesa una provincia en concreto, creamos una variable de dicha provincia

columnas_importantes = ["fecha", "dia_inicio", "dia_cuarentena_dnu260",
                     "nue_casosconf_diff", "nue_fallecidos_diff",
                     "informe_link", "observacion"]
```
